### Ejercicio 1
Definición de histogramas de Leading Jets con y sin peso.

```python
# Set Histograms

# Leading Reco-Jet R4 pT Weighted
hist_lead_reco_R4_pT_evtw = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Leading Reco-Jet R4 pT
hist_lead_reco_R4_pT = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Leading Reco-Jet R10 pT Weighted
hist_lead_reco_R10_pT_evtw = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Leading Reco-Jet R10 pT
hist_lead_reco_R10_pT = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Leading Reco-Jet R10 Trimmed pT Weighted
hist_lead_reco_R10_Trimmed_pT_evtw = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Leading Reco-Jet R10 Trimmed pT
hist_lead_reco_R10_Trimmed_pT = ROOT.TH1F(
    "Leading Reco-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Leading Truth-Jet R4 pT Weighted
hist_lead_truth_R4_pT_evtw = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Leading Truth-Jet R4 pT
hist_lead_truth_R4_pT = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Leading Truth-Jet R10 pT Weighted
hist_lead_truth_R10_pT_evtw = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Leading Truth-Jet R10 pT
hist_lead_truth_R10_pT = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Leading Truth-Jet R10 Trimmed pT Weighted
hist_lead_truth_R10_Trimmed_pT_evtw = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Leading Truth-Jet R10 Trimmed pT
hist_lead_truth_R10_Trimmed_pT = ROOT.TH1F(
    "Leading Truth-jet",
    "Leading jet pT; pT (GeV);Events",
    50,
    10,
    500
)
```

Llenado de los histogramas de Leading Jets.

```python
# Fill Histograms
for event in tree:
    
    # Fill Leading Reco-Jets Histograms
    if(tree.RecoJets_R4_pt.size() > 0):
        hist_lead_reco_R4_pT_evtw.Fill(tree.RecoJets_R4_pt.at(0)/1000, tree.EventWeight)
        hist_lead_reco_R4_pT.Fill(tree.RecoJets_R4_pt.at(0)/1000)
    if(tree.RecoJets_R10_pt.size() > 0):
        hist_lead_reco_R10_pT_evtw.Fill(tree.RecoJets_R10_pt.at(0)/1000, tree.EventWeight)
        hist_lead_reco_R10_pT.Fill(tree.RecoJets_R10_pt.at(0)/1000)
    if(tree.RecoJets_R10_Trimmed_pt.size() > 0):
        hist_lead_reco_R10_Trimmed_pT_evtw.Fill(tree.RecoJets_R10_Trimmed_pt.at(0)/1000, tree.EventWeight)
        hist_lead_reco_R10_Trimmed_pT.Fill(tree.RecoJets_R10_Trimmed_pt.at(0)/1000)
        
    # Fill Leading Truth-Jets Histograms
    if(tree.TruthJets_R4_pt.size() > 0):
        hist_lead_truth_R4_pT_evtw.Fill(tree.TruthJets_R4_pt.at(0)/1000, tree.EventWeight)
        hist_lead_truth_R4_pT.Fill(tree.TruthJets_R4_pt.at(0)/1000)
    if(tree.TruthJets_R10_pt.size() > 0):
        hist_lead_truth_R10_pT_evtw.Fill(tree.TruthJets_R10_pt.at(0)/1000, tree.EventWeight)
        hist_lead_truth_R10_pT.Fill(tree.TruthJets_R10_pt.at(0)/1000)
    if(tree.TruthJets_R10_Trimmed_pt.size() > 0):
        hist_lead_truth_R10_Trimmed_pT_evtw.Fill(tree.TruthJets_R10_Trimmed_pt.at(0)/1000, tree.EventWeight)
        hist_lead_truth_R10_Trimmed_pT.Fill(tree.TruthJets_R10_Trimmed_pt.at(0)/1000)
        
print("¡Listo!")
```

Imprimir y guardar los histogramas.

```python
# Save Histograms

canvas.SetLogy(1)

# Leading Jet R4 pT Weighted
hist_lead_reco_R4_pT_evtw.SetStats(0)
hist_lead_reco_R4_pT_evtw.SetMarkerStyle(20)
hist_lead_reco_R4_pT_evtw.SetMarkerColor(2)
hist_lead_reco_R4_pT_evtw.Draw("")
hist_lead_truth_R4_pT_evtw.SetMarkerStyle(21)
hist_lead_truth_R4_pT_evtw.SetMarkerColor(4)
hist_lead_truth_R4_pT_evtw.Draw("same")
canvas.Print("Leading jet R4 pT Weighted.pdf")
canvas.Clear()

# Leading Jet R4 pT
hist_lead_reco_R4_pT.SetStats(0)
hist_lead_reco_R4_pT.SetFillColorAlpha(2,0.8)
hist_lead_reco_R4_pT.Draw("")
hist_lead_truth_R4_pT.SetFillColorAlpha(4, 0.4)
hist_lead_truth_R4_pT.Draw("same")
canvas.Print("Leading jet R4 pT.pdf")
canvas.Clear()

# Leading Jet R10 pT Weighted
hist_lead_reco_R10_pT_evtw.SetStats(0)
hist_lead_reco_R10_pT_evtw.SetMarkerStyle(20)
hist_lead_reco_R10_pT_evtw.SetMarkerColor(2)
hist_lead_reco_R10_pT_evtw.Draw("")
hist_lead_truth_R10_pT_evtw.SetMarkerStyle(21)
hist_lead_truth_R10_pT_evtw.SetMarkerColor(4)
hist_lead_truth_R10_pT_evtw.Draw("same")
canvas.Print("Leading jet R10 pT Weighted.pdf")
canvas.Clear()

# Leading Jet R10 pT
hist_lead_reco_R10_pT.SetStats(0)
hist_lead_reco_R10_pT.SetFillColorAlpha(2, 0.8)
hist_lead_reco_R10_pT.Draw("")
hist_lead_truth_R10_pT.SetFillColorAlpha(4, 0.4)
hist_lead_truth_R10_pT.Draw("same")
canvas.Print("Leading jet R10 pT.pdf")
canvas.Clear()

# Leading Jet R10 Trimmed pT Weighted
hist_lead_reco_R10_Trimmed_pT_evtw.SetStats(0)
hist_lead_reco_R10_Trimmed_pT_evtw.SetMarkerStyle(20)
hist_lead_reco_R10_Trimmed_pT_evtw.SetMarkerColor(2)
hist_lead_reco_R10_Trimmed_pT_evtw.Draw("")
hist_lead_truth_R10_Trimmed_pT_evtw.SetMarkerStyle(21)
hist_lead_truth_R10_Trimmed_pT_evtw.SetMarkerColor(4)
hist_lead_truth_R10_Trimmed_pT_evtw.Draw("same")
canvas.Print("Leading jet R10 Trimmed pT Weighted.pdf")
canvas.Clear()

# Leading Jet R10 Trimmed pT
hist_lead_reco_R10_Trimmed_pT.SetStats(0)
hist_lead_reco_R10_Trimmed_pT.SetFillColorAlpha(2, 0.8)
hist_lead_reco_R10_Trimmed_pT.Draw("")
hist_lead_truth_R10_Trimmed_pT.SetFillColorAlpha(4, 0.4)
hist_lead_truth_R10_Trimmed_pT.Draw("same")
canvas.Print("Leading jet R10 Trimmed pT.pdf")
canvas.Clear()

canvas.SetLogy(0)
```

### Ejercicio 2
Definición de histogramas de todos los Jets con y sin peso.

```python
# Set Histograms

# Reco-Jet R4 pT Weighted
hist_reco_R4_pT_evtw = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Reco-Jet R4 pT
hist_reco_R4_pT = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Reco-Jet R10 pT Weighted
hist_reco_R10_pT_evtw = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Reco-Jet R10 pT
hist_reco_R10_pT = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Reco-Jet R10 Trimmed pT Weighted
hist_reco_R10_Trimmed_pT_evtw = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Reco-Jet R10 Trimmed pT
hist_reco_R10_Trimmed_pT = ROOT.TH1F(
    "Reco-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Truth-Jet R4 pT Weighted
hist_truth_R4_pT_evtw = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Truth-Jet R4 pT
hist_truth_R4_pT = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Truth-Jet R10 pT Weighted
hist_truth_R10_pT_evtw = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Truth-Jet R10 pT
hist_truth_R10_pT = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    80,
    500
)

# Truth-Jet R10 Trimmed pT Weighted
hist_truth_R10_Trimmed_pT_evtw = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Truth-Jet R10 Trimmed pT
hist_truth_R10_Trimmed_pT = ROOT.TH1F(
    "Truth-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    500
)

# Track-Jet R4 pT Weighted
hist_track_R4_pT_evtw = ROOT.TH1F(
    "Track-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)

# Track-Jet R4 pT
hist_track_R4_pT = ROOT.TH1F(
    "Track-jet",
    "jet pT; pT (GeV);Events",
    50,
    10,
    200
)
```

Llenado de los histogramas de Leading Jets.

```python
# Fill Histograms
for event in tree:

    # Fill Truth-Jets Histograms
    if(tree.RecoJets_R4_pt.size() > 0):
        for tr in range(tree.RecoJets_R4_pt.size()):
            hist_reco_R4_pT_evtw.Fill(tree.RecoJets_R4_pt.at(tr)/1000, tree.EventWeight)
            hist_reco_R4_pT.Fill(tree.RecoJets_R4_pt.at(tr)/1000)
    if(tree.RecoJets_R10_pt.size() > 0):
        for tr in range(tree.RecoJets_R10_pt.size()):
            hist_reco_R10_pT_evtw.Fill(tree.RecoJets_R10_pt.at(tr)/1000, tree.EventWeight)
            hist_reco_R10_pT.Fill(tree.RecoJets_R10_pt.at(tr)/1000)
    if(tree.RecoJets_R10_Trimmed_pt.size() > 0):
        for tr in range(tree.RecoJets_R10_Trimmed_pt.size()):
            hist_reco_R10_Trimmed_pT_evtw.Fill(tree.RecoJets_R10_Trimmed_pt.at(tr)/1000, tree.EventWeight)
            hist_reco_R10_Trimmed_pT.Fill(tree.RecoJets_R10_Trimmed_pt.at(tr)/1000)

    # Fill Truth-Jets Histograms
    if(tree.TruthJets_R4_pt.size() > 0):
        for tr in range(tree.TruthJets_R4_pt.size()):
            hist_truth_R4_pT_evtw.Fill(tree.TruthJets_R4_pt.at(tr)/1000, tree.EventWeight)
            hist_truth_R4_pT.Fill(tree.TruthJets_R4_pt.at(tr)/1000)
    if(tree.TruthJets_R10_pt.size() > 0):
        for tr in range(tree.TruthJets_R10_pt.size()):
            hist_truth_R10_pT_evtw.Fill(tree.TruthJets_R10_pt.at(tr)/1000, tree.EventWeight)
            hist_truth_R10_pT.Fill(tree.TruthJets_R10_pt.at(tr)/1000)
    if(tree.TruthJets_R10_Trimmed_pt.size() > 0):
        for tr in range(tree.TruthJets_R10_Trimmed_pt.size()):
            hist_truth_R10_Trimmed_pT_evtw.Fill(tree.TruthJets_R10_Trimmed_pt.at(tr)/1000, tree.EventWeight)
            hist_truth_R10_Trimmed_pT.Fill(tree.TruthJets_R10_Trimmed_pt.at(tr)/1000)

    # Fill Track-Jets Histograms
    if(tree.TrackJets_R4_pt.size() > 0):
        for tr in range(tree.TrackJets_R4_pt.size()):
            hist_track_R4_pT_evtw.Fill(tree.TrackJets_R4_pt.at(tr)/1000, tree.EventWeight)
            hist_track_R4_pT.Fill(tree.TrackJets_R4_pt.at(tr)/1000)    

print("¡Listo!")
```

Imprimir y guardar los histogramas.

```python
# Save Histograms

canvas.SetLogy(1)

# Jet R4 pT Weighted
hist_reco_R4_pT_evtw.SetStats(0)
hist_reco_R4_pT_evtw.SetMarkerStyle(20)
hist_reco_R4_pT_evtw.SetMarkerColor(2)
hist_reco_R4_pT_evtw.Draw("")
hist_truth_R4_pT_evtw.SetMarkerStyle(21)
hist_truth_R4_pT_evtw.SetMarkerColor(4)
hist_truth_R4_pT_evtw.Draw("same")
hist_track_R4_pT_evtw.SetMarkerStyle(22)
hist_track_R4_pT_evtw.SetMarkerColor(6)
hist_track_R4_pT_evtw.Draw("same")
canvas.Print("Jet R4 pT Weighted.pdf")
canvas.Clear()

# Jet R4 pT
hist_reco_R4_pT.SetStats(0)
hist_reco_R4_pT.SetFillColorAlpha(2,0.8)
hist_reco_R4_pT.Draw("")
hist_truth_R4_pT.SetFillColorAlpha(4, 0.5)
hist_truth_R4_pT.Draw("same")
hist_track_R4_pT.SetFillColorAlpha(6, 0.2)
hist_track_R4_pT.Draw("same")
canvas.Print("Jet R4 pT.pdf")
canvas.Clear()

# Jet R10 pT Weighted
hist_reco_R10_pT_evtw.SetStats(0)
hist_reco_R10_pT_evtw.SetMarkerStyle(20)
hist_reco_R10_pT_evtw.SetMarkerColor(2)
hist_reco_R10_pT_evtw.Draw("")
hist_truth_R10_pT_evtw.SetMarkerStyle(21)
hist_truth_R10_pT_evtw.SetMarkerColor(4)
hist_truth_R10_pT_evtw.Draw("same")
canvas.Print("Jet R10 pT Weighted.pdf")
canvas.Clear()

# Jet R10 pT
hist_reco_R10_pT.SetStats(0)
hist_reco_R10_pT.SetFillColorAlpha(2, 0.8)
hist_reco_R10_pT.Draw("")
hist_truth_R10_pT.SetFillColorAlpha(4, 0.4)
hist_truth_R10_pT.Draw("same")
canvas.Print("Jet R10 pT.pdf")
canvas.Clear()

# Jet R10 Trimmed pT Weighted
hist_reco_R10_Trimmed_pT_evtw.SetStats(0)
hist_reco_R10_Trimmed_pT_evtw.SetMarkerStyle(20)
hist_reco_R10_Trimmed_pT_evtw.SetMarkerColor(2)
hist_reco_R10_Trimmed_pT_evtw.Draw("")
hist_truth_R10_Trimmed_pT_evtw.SetMarkerStyle(21)
hist_truth_R10_Trimmed_pT_evtw.SetMarkerColor(4)
hist_truth_R10_Trimmed_pT_evtw.Draw("same")
canvas.Print("Jet R10 Trimmed pT Weighted.pdf")
canvas.Clear()

# Jet R10 Trimmed pT
hist_reco_R10_Trimmed_pT.SetStats(0)
hist_reco_R10_Trimmed_pT.SetFillColorAlpha(2, 0.8)
hist_reco_R10_Trimmed_pT.Draw("")
hist_truth_R10_Trimmed_pT.SetFillColorAlpha(4, 0.4)
hist_truth_R10_Trimmed_pT.Draw("same")
canvas.Print("Jet R10 Trimmed pT.pdf")
canvas.Clear()

canvas.SetLogy(0)
```
